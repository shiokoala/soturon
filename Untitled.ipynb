```python
import math
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import griddata
```

```python
df=pd.read_csv("rao.csv")
```

```python
# Fn = T{:,1};
# omega = T{:,4};
# Fx = T{:,10};
# Fy = T{:,13};
# Fz = T{:,16};
# Mx = T{:,19};
# My = T{:,22};
# Mz = T{:,25};

Fn = df['Fn']
omega = df['Ï‰_e[rad/s]']
Fx = df['Fx[N/cm]']
Fy = df['Fy[N/cm]']
Fz = df['Fz[N/cm]']
Mx = df['Mx[N/cm]']
My = df['My[N/cm]']
Mz = df['Mz[N/cm]']
```

```python
grid_x, grid_y = np.mgrid[0:0.8:100j, 0:15:100j]
points = pd.concat([Fn,omega],axis=1).values
values = Fx.values
```

```python
grid_z0 = griddata(points, values, (grid_x, grid_y), method='nearest')
grid_z1 = griddata(points, values, (grid_x, grid_y), method='linear')
grid_z2 = griddata(points, values, (grid_x, grid_y), method='cubic')
```

```python
plt.subplot(131)
plt.imshow(grid_z0)
plt.imshow(grid_z2)
plt.title('Nearest')
plt.subplot(132)
plt.imshow(grid_z1)
plt.title('Linear')
plt.subplot(133)
plt.imshow(grid_z2)
plt.title('Cubic')
plt.gcf().set_size_inches(12, 12)
plt.show()
```

```python
for i in range(grid_z2.shape[0]):
    for j in range(grid_z2.shape[1]):
        if(np.isnan(grid_z2[i][j])):
            grid_z2[i][j]=grid_z0[i][j]
```
